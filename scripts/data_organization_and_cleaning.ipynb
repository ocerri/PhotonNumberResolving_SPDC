```python
dfRaw = pd.DataFrame()
for vn in varName:
    aux = eval(vn)
#     print vn, aux.shape
    dfRaw[vn] = eval(vn)
```

```python
renameDic = {'shg_currents': 'shg_current'}
for vn in varName:
    if vn in renameDic.keys():
        continue
    n = vn.replace('_arr', '')
    n = n.replace('_data', '')
    n = n.replace('photon', '')
    n = n.replace('binary', 'thr')
    n = n.replace('total', 'tot')
    n = n.replace('s1', '1')
    n = n.replace('s2', '2')
    renameDic[vn] = n
print renameDic
df = dfRaw.rename(renameDic, axis='columns')
```

```python
# Compute g_2(0)
iMode = {'pnr': 'single', 'thr': 'tot'}
for mode in iMode.keys():
    g2 = df['Ci12_'+mode] * df['Ci_'+iMode[mode]]
    g2 /= df['Ci1_'+mode] * df['Ci2_'+mode]

    factors = ['Ci12_'+mode, 'Ci_'+iMode[mode], 'Ci1_'+mode, 'Ci2_'+mode]
    g2_err = np.zeros_like(g2)
    for n in factors:
        g2_err += np.square(df[n+'_err']/df[n])
    g2_err = g2 * np.sqrt(g2_err)
    df['g2_'+mode] = g2
    df['g2_'+mode+'_err'] = g2_err
```
